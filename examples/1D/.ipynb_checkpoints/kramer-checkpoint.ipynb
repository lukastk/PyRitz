```python
%matplotlib inline
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
import pyritz
import nlopt
```

```python
x1, x2 = -1, 1
Nm, Nq = 40, 80
a0     = np.zeros(Nm)


def LagrangianFW(x, v, Nq):
    fx = -12*(x*x-1)*x
    return (v-fx)*(v-fx)


def LagrangianOM(x, v, Nq):
    fx = -12*(x*x-1)*x
    ff = -12*(x*x-1) - 24*x*x
    return ff + (v-fx)*(v-fx)
```

```python
def optPath(path):    
    def myfunc(x, grad):
        res=path.action(x)
        return res 
    opt = nlopt.opt(nlopt.LN_NEWUOA, np.size(a0))
    opt.set_maxeval(5000)
    opt.set_min_objective(myfunc)
    opt.set_xtol_rel(1e-12)
    aF = opt.optimize(a0)
    return aF

path1  = pyritz.expansion.GalerkinInstanton(x1, x2, LagrangianFW, Nq)
path2  = pyritz.expansion.GalerkinInstanton(x1, x2, LagrangianOM, Nq)
a1, a2 = optPath(path1), optPath(path2)
```

```python
plt.figure(num=None, figsize=(12, 8), dpi=80, facecolor='w', edgecolor='k')
t = np.linspace(-1, 1, 128)
x = path1.position(a0,t)
y = path1.position(a1, t)
z = path2.position(a2, t)

plt.plot(t, x, 'k-', label='Initial', lw=2, alpha=0.6)
plt.plot(t, y, '-', color="#348ABD", lw=2, label='Freidlin-Wentzell')
plt.plot(t, z, '-', color="#A60628", lw=2, label='Onsager-Machlup')
plt.legend(loc='best', fontsize=18); plt.grid()
```
