```python
# domain
x_start, x_end = 0, -.25

# Chebyshev modes
N = 8
d = 1
path = CollocationInstanton(x_start, x_end, lagrangian, N)
m0 = CollocationInstanton.get_straight_line_path(x_start, x_end, N)

ub = np.full(d*N, np.inf)
lb = np.full(d*N, -np.inf)

ub[0] = x_start
lb[0] = x_start
ub[N-1] = x_end
lb[N-1] = x_end

def get_action(m, grad):
    return path.action(m)

opt = nlopt.opt(nlopt.LN_NELDERMEAD, np.size(m0))
opt.set_lower_bounds(lb)
opt.set_upper_bounds(ub)
opt.set_maxeval(10000)
opt.set_min_objective(get_action)
opt.set_xtol_rel(1e-12)
m = opt.optimize(m0)
```

```python
# plotting
t = np.linspace(-1,1,32)
p = path.get_interpolator(m)
y = p.compute_position(t)

plt.plot(t, y, 'r--', label='bfgs')
plt.legend(loc='upper right');
plt.show()
```

```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
import pyritz

# domain
x1, x2 = 0, -0.25

# Chebyshev modes
Nm = 8
a0 = np.zeros(Nm)

# Lagrangian for the Brachistochrone problem
def Lagrangian(x, v, Nq):
    lg = 0.5*np.sqrt((1+v*v)/(2*10*np.abs( x1 - x )))
    return lg

# instantiate the Instanton class
path = pyritz.expansion.GalerkinInstanton(x1, x2, Lagrangian)
path1 = pyritz.expansion.GalerkinInstanton(x1, x2, Lagrangian)

#minimize the action 
res = minimize(path.action, a0, method='bfgs',options={'gtol': 1e-2, 'disp': True})
res1 = minimize(path1.action, a0, method='bfgs',options={'gtol': 1e-2, 'disp': True})

# plotting
t = np.linspace(-1,1,32)
x = path.position(a0,t)
y = path.position(res.x, t)
z = path1.position(res1.x, t)

plt.plot(t, x, label='initial')
plt.plot(t, y, 'r--', label='bfgs')
plt.plot(t, z, 'co', label='bfgs-2')
plt.legend(loc='upper right');
plt.show()
```
