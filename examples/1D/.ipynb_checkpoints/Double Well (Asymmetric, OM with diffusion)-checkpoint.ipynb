```python
%matplotlib inline
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
import matplotlib as mpl
from pyritz.expansion import CollocationInstanton
import nlopt
from scipy.linalg import null_space
```

```python
r1 = -1
r2 = 1
c = 10
T = 2

k1 = 5
k2 = 10

R = 5

def Upot(x):
    return (x*(12*(k1 - k2) - 6*(k1 + k2)*x + 4*(-k1 + k2)*x**2 + 3*(k1 + k2)*x**3))/48.

def force(x):
    return -( 2*c*(-r1 + x)**2 * (-r2 + x) + 2*c*(-r1 + x)*(-r2 + x)**2 )

def lagrangian_FW(x, v, Nq):
    dv =  2*c*(-r1 + x)**2 * (-r2 + x) + 2*c*(-r1 + x)*(-r2 + x)**2
    lg = (T/4) * ( (2/T)*v + dv )**2
    
    return lg

def lagrangian(x, v, Nq):
    return T * (np.pi**2)/(4 * R**2)   +     (T*(-(x*(8*(-k1 + k2) + 18*(k1 + k2)*x))/48. +       (6*(k1 + k2) - 8*(-k1 + k2)*x - 9*(k1 + k2)*x**2)/24.))/4. + (T*((2*v)/T + (x*(-6*(k1 + k2) + 8*(-k1 + k2)*x + 9*(k1 + k2)*x**2))/48. + (12*(k1 - k2) - 6*(k1 + k2)*x + 4*(-k1 + k2)*x**2 + 3*(k1 + k2)*x**3)/48.)**2)/4.
```

```python
x_start = r1
x_end = r2

R = 2

N = 20
Nq = N*10
d = np.size(x_start)

path = CollocationInstanton(x_start, x_end, lagrangian, N)
m0 = CollocationInstanton.get_straight_line_path(x_start, x_end, N)

T0 = 2
m0 = np.concatenate( (m0, [T0]) )

def get_action(m, grad):
    global T
    T = m[-1]
    print(T)
    return path.action_extra(m[:-1], Nq)

ub = np.full(d*N+1, np.inf)
lb = np.full(d*N+1, -np.inf)

ub[0] = x_start
lb[0] = x_start
ub[N-1] = x_end
lb[N-1] = x_end
lb[N] = 0

opt = nlopt.opt(nlopt.LN_BOBYQA, np.size(m0))
opt.set_lower_bounds(lb)
opt.set_upper_bounds(ub)
opt.set_min_objective(get_action)

#opt.set_maxeval(7000)
opt.set_xtol_abs(1e-20)
m = opt.optimize(m0)

print("Si:", path.action_extra(m0[:-1], Nq))
print(" S:", path.action_extra(m[:-1], Nq))
print(" T:", T)
```

```python
t = np.linspace(-1,1,1000)

p = path.get_interpolator(m[:-1])
x = p.compute_position(t)
pi = path.get_interpolator(m0[:-1])
xi = pi.compute_position(t)

plt.plot(t, x, '-', color="#A60628", lw=2, label='final')
#plt.plot(CollocationInstanton.get_chebyshev_grid(N), m, '.')

plt.plot(t, xi, 'k-', label='initial', lw=2, alpha=0.6)

plt.xlabel("$t$", fontsize=22); plt.ylabel("$x$", fontsize=22)
plt.legend(loc='upper left'); plt.grid(); plt.xlim(-1.2,1.2)
```

#### Iterative

```python
x_start = r1
x_end = r2

T0 = 1
R = 1
d = np.size(x_start)

Nstart = 20
Nend = 70
Nstep = 0.5

m = CollocationInstanton.get_straight_line_path(x_start, x_end, Nstart)
m = np.concatenate( (m, [T0]) )
arch = [(m[:-1], m[-1], Nstart)]

for N in range(Nstart, Nend+1, Nstep):
    Nq = N*10
    path = CollocationInstanton(x_start, x_end, lagrangian, N)

    T = m[-1]
    p = path.get_interpolator(m[:-1])
    t = CollocationInstanton.get_chebyshev_grid(N)
    m = p.compute_position(t)
    m = np.concatenate( (m, [T]) )
    
    def get_action(m, grad):
        global T
        T = m[-1]
        return path.action_extra(m[:-1], Nq)

    ub = np.full(d*N+1, np.inf)
    lb = np.full(d*N+1, -np.inf)

    ub[0] = x_start
    lb[0] = x_start
    ub[N-1] = x_end
    lb[N-1] = x_end
    lb[N] = 0

    opt = nlopt.opt(nlopt.LN_NEWUOA, np.size(m))
    opt.set_lower_bounds(lb)
    opt.set_upper_bounds(ub)
    opt.set_min_objective(get_action)

    opt.set_maxeval(10000)
    #opt.set_xtol_abs(1e-20)
    m = opt.optimize(m)
    
    arch.append( (m[:-1], m[-1], N) )

    print("N:", N)
    print("-  S:", path.action_extra(m[:-1], Nq))
    print("-  T:", m[-1])
```

```python
m = arch[-1][0]

t = np.linspace(-1,1,1000)

p = path.get_interpolator(m)
x = p.compute_position(t)

plt.plot(t, x, '-', color="#A60628", lw=2, label='final')
#plt.plot(CollocationInstanton.get_chebyshev_grid(N), m, '.')

plt.xlabel("$t$", fontsize=22); plt.ylabel("$x$", fontsize=22)
plt.legend(loc='upper left'); plt.grid(); plt.xlim(-1.2,1.2)
```
