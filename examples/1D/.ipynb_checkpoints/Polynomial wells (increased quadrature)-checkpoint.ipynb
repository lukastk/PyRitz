```python
%matplotlib inline
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
import matplotlib as mpl
from pyritz.expansion import CollocationInstanton
import nlopt
```

```python
rs = [1, 4]
C = 2

def u_pot(x):
    return C * (x - rs[0])**2 * (x - rs[1])**2

def force(x):
    return 2* (rs[0]+rs[1] - 2 * x)*(rs[0]-x)*(-rs[1]+x) * C

def dforce(x):
    return (  2*(-rs[0] + x)**2 + 8*(-rs[0] + x)*(-rs[1]+x) + 2*(-rs[1]+x)**2  ) * C

def lagrangian(x, v, Nq):
    f = 2* (rs[0]+rs[1] - 2 * x)*(rs[0]-x)*(-rs[1]+x) * C
    
    return 0.5*(v - f)**2
```

```python
x = np.linspace(0, 5, 1000)
plt.plot(x, u_pot(x))
plt.xlabel("$x$", fontsize=22); plt.ylabel("$U(x)$", fontsize=22)

#plt.plot(x, 2* (rs[0]+rs[1] - 2 * x)*(rs[0]-x)*(-rs[1]+x) * C)
#plt.xlabel("$x$", fontsize=22); plt.ylabel("$U(x)$", fontsize=22)
```

```python
x_start = rs[0]
x_end = rs[1]

N = 40
Nq = 1000
d = 1

path = CollocationInstanton(x_start, x_end, lagrangian, N)

m0 = CollocationInstanton.get_straight_line_path(x_start, x_end, N)

ub = np.full(d*N, np.inf)
lb = np.full(d*N, -np.inf)

ub[0] = x_start
lb[0] = x_start
ub[N-1] = x_end
lb[N-1] = x_end

def get_action(m, grad):
    return path.action_extra(m, Nq)

opt = nlopt.opt(nlopt.LN_NEWUOA, np.size(m0))
opt.set_lower_bounds(lb)
opt.set_upper_bounds(ub)
opt.set_maxeval(1000)
opt.set_min_objective(get_action)
opt.set_xtol_rel(1e-12)
m = opt.optimize(m0)

print("S = ", path.action_extra(m, Nq))
```

```python
t = np.linspace(-1,1,1000)

p = path.get_interpolator(m)
x = p.compute_position(t)
pi = path.get_interpolator(m0)
xi = pi.compute_position(t)

plt.plot(t, x, '-', color="#A60628", lw=2, label='final')
plt.plot(t, xi, 'k-', label='initial', lw=2, alpha=0.6)
#plt.plot(CollocationInstanton.get_chebyshev_grid(N), m, ".")

plt.xlabel("$t$", fontsize=22); plt.ylabel("$x$", fontsize=22)
plt.legend(loc='upper left'); plt.grid(); plt.xlim(-1.2,1.2)
```

```python
def get_path():
    x_start = rs[0]
    x_end = rs[1]
    d = 1
    
    path = CollocationInstanton(x_start, x_end, lagrangian, N)

    m0 = CollocationInstanton.get_straight_line_path(x_start, x_end, N)

    ub = np.full(d*N, np.inf)
    lb = np.full(d*N, -np.inf)

    ub[0] = x_start
    lb[0] = x_start
    ub[N-1] = x_end
    lb[N-1] = x_end

    def get_action(m, grad):
        return path.action(m)

    opt = nlopt.opt(nlopt.LN_NEWUOA, np.size(m0))
    opt.set_lower_bounds(lb)
    opt.set_upper_bounds(ub)
    opt.set_maxeval(1000)
    opt.set_min_objective(get_action)
    opt.set_xtol_rel(1e-12)
    m = opt.optimize(m0)
    return m

Cstart = 0
Cend = 2
Cstep = 0.01

res = []
cs = []

N = 30

for i in range(int((Cend-Cstart)/Cstep)):
    C = Cstart + Cstep*i
    cs.append(C)
    m = get_path()
    
    path2 = CollocationInstanton(x_start, x_end, lagrangian, N)
    S1 = path2.action(m)
    
    p = path2.get_interpolator(m)
    path2 = CollocationInstanton(x_start, x_end, lagrangian, 300)
    cheb_t = CollocationInstanton.get_chebyshev_grid(N2)
    S2 = path2.action(p.compute_position(cheb_t))
    res.append(S2 - S1)
    
plt.plot(cs, res)
```

#### Annealing

```python
x_start = rs[0]
x_end = rs[1]

N = 10
d = 1

Cstart = 0.1
Cend = 5
Cstep = 0.01
d = 1

m0 = CollocationInstanton.get_straight_line_path(x_start, x_end, N)

for i in range(int((Cend-Cstart)/Cstep)):
    C = Cstart + Cstep*i
    
    path = CollocationInstanton(x_start, x_end, lagrangian, N)
    
    ub = np.full(d*N, np.inf)
    lb = np.full(d*N, -np.inf)

    ub[0] = x_start
    lb[0] = x_start
    ub[N-1] = x_end
    lb[N-1] = x_end

    def get_action(m, grad):
        return path.action(m)

    opt = nlopt.opt(nlopt.LN_NEWUOA, np.size(m0))
    opt.set_lower_bounds(lb)
    opt.set_upper_bounds(ub)
    opt.set_maxeval(1000)
    opt.set_min_objective(get_action)
    opt.set_xtol_rel(1e-12)
    m0 = opt.optimize(m0)
    
m = m0
    
print("S =", path.action(m))
```

```python
t = np.linspace(-1,1,1000)

p = path.get_interpolator(m)
x = p.compute_position(t)

plt.plot(t, x, '-', color="#A60628", lw=2, label='final')
#plt.plot(t, exact_sol(t), '--', label='exact')

plt.xlabel("$t$", fontsize=22); plt.ylabel("$x$", fontsize=22)
plt.legend(loc='upper left'); plt.grid(); plt.xlim(-1.2,1.2)
```

#### Iterative approach

```python
x_start = rs[0]
x_end = rs[1]

Nstart = 4
Nend = 50
Nq = 200
d = 1

m0 = CollocationInstanton.get_straight_line_path(x_start, x_end, N)

for N in range(Nstart, Nend):
    cheb_t = CollocationInstanton.get_chebyshev_grid(N)
    path = CollocationInstanton(x_start, x_end, lagrangian, N)

    p = path.get_interpolator(m)
    m0 = p.compute_position(cheb_t)

    ub = np.full(d*N, np.inf)
    lb = np.full(d*N, -np.inf)

    ub[0] = x_start
    lb[0] = x_start
    ub[N-1] = x_end
    lb[N-1] = x_end

    def get_action(m, grad):
        return path.action_extra(m, Nq)

    opt = nlopt.opt(nlopt.LN_NEWUOA, np.size(m0))
    opt.set_lower_bounds(lb)
    opt.set_upper_bounds(ub)
    opt.set_maxeval(1000)
    opt.set_min_objective(get_action)
    opt.set_xtol_rel(1e-12)
    m = opt.optimize(m0)
    
print("S =", path.action_extra(m, Nq))
```

```python
t = np.linspace(-1,1,1000)

p = path.get_interpolator(m)
x = p.compute_position(t)

plt.plot(t, x, '-', color="#A60628", lw=2, label='final')
#plt.plot(t, exact_sol(t), '--', label='exact')

plt.xlabel("$t$", fontsize=22); plt.ylabel("$x$", fontsize=22)
plt.legend(loc='upper left'); plt.grid(); plt.xlim(-1.2,1.2)
```

```python
res = []

Nstart = 50
Nend = 600

for N2 in range(Nstart, Nend + 1):
    t = CollocationInstanton.get_chebyshev_grid(N2)
    p = path.get_interpolator(m)
    cheb_t = CollocationInstanton.get_chebyshev_grid(N2)
    path2 = CollocationInstanton(x_start, x_end, lagrangian, N2)
    res.append( path2.action(p.compute_position(t)) )
    
plt.xlabel("N")
plt.ylabel("S")
plt.plot(np.arange(Nstart, Nend + 1), res)

print("N =", Nstart, "   S =", res[0])
print("N =",Nend, "   S =", res[-1])
```
