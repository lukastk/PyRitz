```python
%matplotlib inline
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
import matplotlib as mpl
from pyritz.expansion import CollocationInstanton
import nlopt
from scipy.linalg import null_space
```

```python
r1 = -1
r2 = 1
c = 10
T = 2

k1 = 10
k2 = 11

def Upot(x):
    return (x*(12*(k1 - k2) - 6*(k1 + k2)*x + 4*(-k1 + k2)*x**2 + 3*(k1 + k2)*x**3))/48.

def force(x):
    return -( 2*c*(-r1 + x)**2 * (-r2 + x) + 2*c*(-r1 + x)*(-r2 + x)**2 )

def lagrangian_FW(x, v, Nq):
    dv =  2*c*(-r1 + x)**2 * (-r2 + x) + 2*c*(-r1 + x)*(-r2 + x)**2
    lg = (T/4) * ( (2/T)*v + dv )**2
    
    return lg

def lagrangian(x, v, Nq):
    return         (T*(-(x*(8*(-k1 + k2) + 18*(k1 + k2)*x))/48. +       (6*(k1 + k2) - 8*(-k1 + k2)*x - 9*(k1 + k2)*x**2)/24.))/4. + (T*((2*v)/T + (x*(-6*(k1 + k2) + 8*(-k1 + k2)*x + 9*(k1 + k2)*x**2))/48. + (12*(k1 - k2) - 6*(k1 + k2)*x + 4*(-k1 + k2)*x**2 + 3*(k1 + k2)*x**3)/48.)**2)/4.
```

```python
t = np.linspace(-1.5, 1.5, 1000)
plt.plot(t, Upot(t))
plt.xlabel("x")
plt.ylabel("U(x)")
```

```python
x_start = r1
x_end = r2

N = 50
Nq = N*10
d = np.size(x_start)
T = 100
```

```python
path = CollocationInstanton(x_start, x_end, lagrangian, N)

m0 = CollocationInstanton.get_straight_line_path(x_start, x_end, N)

def get_action(m, grad):
    return path.action_extra(m, Nq)

ub = np.full(d*N, np.inf)
lb = np.full(d*N, -np.inf)

ub[0] = x_start
lb[0] = x_start
ub[N-1] = x_end
lb[N-1] = x_end

opt = nlopt.opt(nlopt.LN_NEWUOA, np.size(m0))
opt.set_lower_bounds(lb)
opt.set_upper_bounds(ub)
opt.set_min_objective(get_action)

opt.set_maxeval(7000)
#opt.set_xtol_abs(1e-20)
m = opt.optimize(m0)

print("Si:", path.action_extra(m0, Nq))
print(" S:", path.action_extra(m, Nq))
```

```python
arch = []

Tstart = 0.1
Tend = 12
Tstep = 0.2

x_start = r1
x_end = r2

N = 40
Nq = N*10
d = np.size(x_start)

for Ts in np.arange(Tstart, Tend + Tstep, Tstep):
    T = Ts

    path = CollocationInstanton(x_start, x_end, lagrangian, N)

    m0 = CollocationInstanton.get_straight_line_path(x_start, x_end, N)

    def get_action(m, grad):
        return path.action_extra(m, Nq)

    ub = np.full(d*N, np.inf)
    lb = np.full(d*N, -np.inf)

    ub[0] = x_start
    lb[0] = x_start
    ub[N-1] = x_end
    lb[N-1] = x_end

    opt = nlopt.opt(nlopt.LN_NEWUOA, np.size(m0))
    opt.set_lower_bounds(lb)
    opt.set_upper_bounds(ub)
    opt.set_min_objective(get_action)

    opt.set_maxeval(10000)
    #opt.set_xtol_abs(1e-20)
    m = opt.optimize(m0)
    
    arch.append( (m, Ts) )
    print(Ts)
```

```python
def group(seq, sep):
    g = []
    for el in seq:
        if el == sep:
            yield g
            g = []
        g.append(el)
    yield g

x1 = -1.1069207893499804
x2 = 1.0972686028420022
r = 0.1
 
#x1 = -1
#x2 = 1
#r = 0.25
    
times = []
    
for m, T in arch:
    t = CollocationInstanton.get_chebyshev_grid(1000)
    p = path.get_interpolator(m)
    x = p.compute_position(t)

    xs1 = [ (tt, xx) if np.abs(xx - x1) < r else None for xx, tt in zip(x, t)]
    xs2 = [ (tt, xx) if np.abs(xx - x2) < r else None for xx, tt in zip(x, t)]

    c = [l for l in list(group(xs1, None)) if (l != [] and len(l) != 1)]
    c = [l[1:] for l in c]
    c = [ l[-1][0] - l[0][0] for l in c ]
    ts1 = c

    c = [l for l in list(group(xs2, None)) if (l != [] and len(l) != 1)]
    c = [l[1:] for l in c]
    c = [ l[-1][0] - l[0][0] for l in c ]
    ts2 = c
    
    t1, t2 = np.NAN, np.NAN
    if len(ts1) == 1:
        t1 = ts1[0]
    elif len(ts1) > 1:
        print("error:", ts1)
        
    if len(ts2) == 1:
        t2 = ts2[0]
    elif len(ts2) > 1:
        print("error", ts1)
        
    t1 = 0.5 * T * t1
    t2 = 0.5 * T * t2
        
    times.append( (t1, t2, float(t2)/t1) )
```

```python
x1 = -1.1069207893499804
x2 = 1.0972686028420022
r = 0.1
 
#x1 = -1
#x2 = 1
#r = 0.25
    
times = []
    
for m, T in arch:
    t = np.linspace(-1, 1, 10000)
    dt = 2.0/10000
    p = path.get_interpolator(m)
    x = p.compute_position(t)

    x_min = x1 - r
    x_max = x1 + r
    mask = (x > x_min)*(x < x_max)
    N = np.sum(mask)
    T1 = N*dt
    
    x_min = x2 - r
    x_max = x2 + r
    mask = (x > x_min)*(x < x_max)
    N = np.sum(mask)
    T2 = N*dt
    
    T1 = 0.5 * T * T1
    T2 = 0.5 * T * T2
    
    if T1 == 0:
        T1T2 = np.nan
    else:
        T1T2 = float(T2)/T1
    
    times.append( (T1, T2, T1T2) )
```
