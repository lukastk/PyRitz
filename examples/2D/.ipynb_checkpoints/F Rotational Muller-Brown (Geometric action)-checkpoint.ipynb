# Rotational Muller-Brown

**Lagrangian:** Geometric Action
<br>
**Implementation:** Python

**System:**

Let $\mathbf{r}$ and $\theta$ be the positional and rotation coordinates respectively, and let $\mathbf{x} = (\mathbf{r}, \theta)$. The Muller-Brown drift term is

$$
\begin{aligned}
U(\mathbf{r}) & = \sum_{i=1}^4 u^{(i)} \\
u^{(i)} & = A^{(i)} \exp \left\{ (\mathbf{r} - \mathbf{r}^{(i)})^T M^{(i)} (\mathbf{r} - \mathbf{r}^{(i)}) \right\} \\
\mathbf{F}(\mathbf{r}) & = - \nabla U
\end{aligned}
$$

and the system is defined as

$$\begin{aligned}
d\mathbf{r} & = \frac{1}{\gamma} \mathbf{F}dt + \frac{1}{\gamma} \nu_s \mathbf{p} + \frac{1}{\gamma} \sigma d\mathbf{W} \\
d\theta & = \frac{1}{\gamma_R} \sigma_R\ d W_R
\end{aligned}
$$

where

$$
\mathbf{p} = (\cos \theta, \sin \theta).
$$

Usually, we set $\gamma$, $\gamma_R$, $\sigma$ and $\sigma_R$ such that

$$
\begin{aligned}
\gamma & = \gamma_R = 1 \\
\frac{1}{\gamma} \sigma & = \frac{1}{\gamma_R} \sigma_R = I
\end{aligned}
$$

## System

### Parameters

System parameters are defined here. All system parameter variables are prefaced with "m_", with the exception of `dim` which is the dimension of the system.

```python
dim = 3
rdim = 2

m_kT = 0.5

m_gamma = 1
m_rgamma = 1/m_gamma
m_sigma = np.sqrt(2 * m_kT / m_gamma)
m_rsigma = (1/m_sigma)
m_rsigma_sqrd = m_rsigma**2
m_nu_s = 1

m_gamma_R = 1
m_rgamma_R = 1/m_gamma_R
m_sigma_R = np.sqrt(2 * m_kT / m_gamma_R)
m_rsigma_R = (1/m_sigma_R)
m_rsigma_R_sqrd = m_rsigma_R**2

m_As = np.array([-200, -100, -170, 15])
m_a = np.array([-1, -1, -6.5, .7])
m_b = np.array([0, 0, 11, .6])
m_c = np.array([-10, -10, -6.5, .7])
m_Ms = np.array([[[-1, 0], [0, -10]],
    [[-1, 0], [0, -10]],
    [[-6.5, 11./2], [11./2, -6.5]],
    [[0.7, 0.6/2], [0.6/2, 0.7]]])
m_ss = np.array([[1, 0], [0, 0.5], [-0.5, 1.5], [-1, 1]])
```

### Dynamics

`system_a` is the drift term of the system. It's used for plotting stream-plots, as well as finding fixed points of the deterministic system.

`lagrangian` is the Lagrangian of the system. The Lagrangian used is the Geometric Action lagrangian (Heymann and Vanden-Eijnden), defined as

$$
L(x, \dot x) = |\dot x|_D |a(x)|_D - \langle \dot x , a(x) \rangle_D
$$

for any general Ito diffusion of the form

$$
dX = a(X)dt + \sqrt{\epsilon} dW.
$$

where $D=\sigma \sigma^T$ and the inner product is defined as $\langle a, b \rangle_D = \langle a, D^{-1} b \rangle$.The function also computes the partial derivatives of the Lagrangian, which are used for gradient optimisation. This can be enabled/disabled by setting the *args* argument to True/False. The partial derivatives of the geometric lagrangian are

$$\begin{aligned}
\frac{\partial L}{\partial x} & =  \nabla a( x ) \cdot \left( |\dot x|_D \hat{a}(x) - \dot x \right) \\
\frac{\partial L}{\partial \dot x} & = |a(x)|_D \hat{\dot{x}} - a(x)
\end{aligned}$$

An important property of the geometric action is that it is time-reparameterisation invariant, meaning that $x$ can be replaced by any other path that shares the same *graph*.

#### Notes on the Muller-Brown system

In this section some useful derivatives are computed, used for the gradient of the geometric action. Some relevant derivatives are

$$
\begin{aligned}
\frac{\partial u^{(k)}}{\partial x_i} & = 2 u^{(k)}(\mathbf{x}) \left[ M^{(k)}_{ij}(\mathbf{x}_j - \mathbf{r}_j^{(k)})\right] \\
\frac{ \partial^2 u^{(k)} }{\partial x_i \partial x_j} & = 2 u^{(k)}(\mathbf{x}) \left[ 2 M_{il}^{(k)} (x_l - r_l^{(k)}) M_{jn}^{(k)} (x_n - r_n^{(k)}) + M_{ji}^{(k)} \right] \\
\frac{ \partial^2 U}{\partial x_i \partial x_j} & = \sum_{k=1}^4 2 u^{(k)}(\mathbf{x}) \left[ 2 M_{il}^{(k)} (x_l - r_l^{(k)}) M_{jn}^{(k)} (x_n - r_n^{(k)}) + M_{ji}^{(k)} \right]
\end{aligned}
$$

We have $U(\mathbf{x}) = \sum_{i=1}^4 u^{(i)}$ and $F = - \nabla_r U$, so

$$\begin{aligned}
    \nabla_r F & = -\sum_{k=1}^4 2 u^{(k)}(\mathbf{x}) \left[ 2 (M^{(k)} (x - r^{(k)}))\ (M^{(n)} (x - r^{(k)}))^T + M^{(k)} \right].
\end{aligned}$$

The drift of the entire system (position and rotation) is $a = \left( \frac{1}{\gamma} \mathbf{F} + \frac{1}{\gamma} \nu_s \mathbf{p}, 0 \right)$. The second term in the positional component has the derivative

$$
\partial_\theta p = (-\sin \theta, \cos \theta).
$$

Letting $\nabla = (\partial_1, \partial_2, \partial_\theta)$, we can write the derivative of the drift term as

$$
\nabla a = \begin{pmatrix}
\partial_1 F_1 & \partial_1 F_2 & 0 \\
\partial_2 F_1 & \partial_2 F_2 & 0 \\
-\sin \theta & \cos \theta & 0
\end{pmatrix}
$$

```python
def system_a(x, y):
    xs = np.array([x,y])
    rs = np.zeros((2, 4))
    us = np.zeros(4)
    a_s = np.zeros(2)
    
    for i in range(4):
        rs[:, i] = m_Ms[i, :, :].dot( xs[:] - m_ss[i, :] )
        us[i] = m_As[i]*np.exp( (xs[:] - m_ss[i, :]).dot(rs[:, i]) )

        a_s[:] += -2*us[i]*rs[:, i]
        
    return a_s

def system_pot(x,y):
    pot=np.zeros(x.shape)
    _a = m_Ms[:, 0, 0]
    _b = m_Ms[:, 0, 1]*2
    _c = m_Ms[:, 1, 1]
    
    for i in range(4):
        dd = _a[i]*(x-m_ss[i][0])**2 + _b[i]*(x-m_ss[i][0])*(y-m_ss[i][1]) + _c[i]*(y-m_ss[i][1])**2
        pot += m_As[i]*np.exp(dd) 
        
    return pot

def lagrangian(ls, dxls, dvls, fvals, ts, args):
    compute_gradient = args
    
    xs, dxs = fvals
    
    rxs = xs[:, :2]
    drxs = dxs[:, :2]
    thetas = xs[:, 2]
    dthetas = dxs[:, 2]
    
    rs = np.zeros((rdim, 4, len(ts)))
    us = np.zeros((4, len(ts)))
    a_s = np.zeros((dim, len(ts)))
    
    _a = m_Ms[:, 0, 0]
    _b = m_Ms[:, 0, 1]*2
    _c = m_Ms[:, 1, 1]

    for i in range(4):
        us[i, :] = m_As[i]*np.exp(
            _a[i]*(rxs[:, 0]-m_ss[i][0])**2 + _b[i]*(rxs[:, 0]-m_ss[i][0])*(rxs[:, 1]-m_ss[i][1]) +
            _c[i]*(rxs[:, 1]-m_ss[i][1])**2)
        rs[0, i, :] = _a[i]*(rxs[:, 0]-m_ss[i][0])+0.5*_b[i]*(rxs[:, 1]-m_ss[i][1])
        rs[1, i, :] = _c[i]*(rxs[:, 1]-m_ss[i][1])+0.5*_b[i]*(rxs[:, 0]-m_ss[i][0])
        a_s[0, :] += -2*us[i, :]*rs[0, i, :]
        a_s[1, :] += -2*us[i, :]*rs[1, i, :]

    a_s[0, :] += m_nu_s*np.cos(thetas)
    a_s[1, :] += m_nu_s*np.sin(thetas)
    
    a_s[:, :] *= m_rgamma
    
    dxs_norm = m_rsigma*np.linalg.norm(dxs, axis=1)
    a_norm = m_rsigma*np.linalg.norm(a_s, axis=0)
    dxs_dot_a = m_rsigma_sqrd*(a_s[0,:].T*drxs[:,0] + a_s[1,:].T*drxs[:, 1] + 0)
    
    ls[:] = dxs_norm * a_norm - dxs_dot_a
    
    # Compute gradient
    
    if compute_gradient:

        ## Compute dL/dx

        da = np.zeros( (dim, dim, len(ts)) )

        for j in range(len(ts)):
            for i in range(4):
                da[:2, :2, j] += - 2 * us[i, j] * (2*np.outer(rs[:, i, j], rs[:, i, j]) + m_Ms[i]) 

        da[:2, :2, :] *= m_rgamma
                
        da[2, 0, :] = -m_nu_s*np.sin(thetas)
        da[2, 1, :] = m_nu_s*np.cos(thetas)
                
        a_normalized = a_s/a_norm

        a_normalized[:, np.where(a_norm==0)] = 0 # Set zero-norm vectors to 0

        for i in range(dxls.shape[0]):
            dxls[i, :] = da[:,:,i].dot( dxs_norm[i]*a_normalized[:,i] - dxs[i,:])

        ## Compute dL/dv

        dxs_2 = (dxs.T*(a_norm/dxs_norm)).T
        dxs_2[ np.where(dxs_norm==0),:] = 0

        dvls[:] = dxs_2 - a_s.T
```

### Find fixed points of the system

```python
from scipy.optimize import root

e_xa  = root(lambda x : system_a(x[0], x[1]), np.array([-0.5, 1.5])).x
e_xb  = root(lambda x : system_a(x[0], x[1]), np.array([0, 0.4])).x
e_xc  = root(lambda x : system_a(x[0], x[1]), np.array([0.5, 0])).x
e_xs1  = root(lambda x : system_a(x[0], x[1]), np.array([-0.7, 0.5])).x
e_xs2 = root(lambda x : system_a(x[0], x[1]), np.array([0.2, 0.4])).x

print(e_xa, e_xb, e_xc, e_xs1, e_xs2)
```

## Optimisation

### Gradient-free optimisation

```python
m_nu_s = 200

x_start = np.concatenate( [e_xa, [0]])
x_end = np.concatenate( [[0.6,0.16], [0]])

Nm = 4
Nq = Nm*10

m0_path = pyritz.funcs.CollocationFF.get_straight_line_path(x_start, x_end, Nm,
                                    exclude_start_point=True, exclude_end_point=True)
m0_endpoint_rot = np.array([x_start[-1], x_end[-1]])
m0 = np.concatenate([m0_endpoint_rot, m0_path])

ff = pyritz.funcs.CollocationFF(Nm, dim, derivatives=1,
                               fixed_start_point=x_start, fixed_end_point=x_end)

quad_scheme = pyritz.quads.Q_clenshaw_curtis

compute_gradient=False
act = pyritz.Action(dim, ff, lagrangian, Nq, quad_scheme,
                    lagrangian_args=compute_gradient)

def get_path(m):
    m_path = m[2:]
    x_start[-1] = m[0]
    x_end[-1] = m[1]
    
    ff.set_fixed_start_point(x_start)
    ff.set_fixed_end_point(x_end)
    return m_path

def get_action(m, grad):
    return act.compute(get_path(m))

opt = nlopt.opt(nlopt.LN_NEWUOA, np.size(m0))
opt.set_min_objective(get_action)
opt.set_xtol_rel(1e-12)
opt.set_maxeval(5000)
m = opt.optimize(m0)

print(act.compute(get_path(m0)))
print(act.compute(get_path(m)))
```

#### Instanton plot

```python
# Plot path

ts = np.linspace(-1, 1, 1000)

paths = [
    (m0, "Initial", "blue"),
    (m, "Final", "green")
]

for p in paths:
    _m, _mlabel, _color = p
    _m = get_path(_m)
    xs, vs = ff.evaluate(_m, ts)
    x, y, theta = xs[:, 0], xs[:, 1], xs[:, 2]
    px, py = np.array([np.cos(theta), np.sin(theta)])/10
    
    plt.plot(x, y, label=_mlabel, color=_color)
    
    arrow_step = 30
    for i in range(len(px)):
        if i%arrow_step==0:
            plt.plot( [x[i],x[i]+px[i]], [y[i],y[i]+py[i]], color="black", alpha=0.6)

# Plot contour of potential

Nq=128
t=np.linspace(-1,1, Nq)
X, Y = np.meshgrid(np.linspace(-1.5, 1.2, 128), np.linspace(-.5, 2, 128))
plt.contour(X, Y, system_pot(X, Y), 256, cmap='gray', linewidths=1, alpha=0.4);

# Plot fixed points

plt.plot(e_xa[0], e_xa[1], 'o', mfc='#AA0000', mec='#AA0000', ms=4)
plt.plot(e_xb[0], e_xb[1], 'o', mfc='#AA0000', mec='#AA0000', ms=4)
plt.plot(e_xc[0], e_xc[1], 'o', mfc='#AA0000', mec='#AA0000', ms=4)
plt.plot(e_xs1[0], e_xs1[1], 'o', mfc='#00AA22', mec='#00AA22', ms=4)
plt.plot(e_xs2[0], e_xs2[1], 'o', mfc='#00AA22', mec='#00AA22', ms=4)

# Plot limits and setup

plt.xlim(-1.3, 1.2);plt.ylim(-.5, 2);
plt.xlabel("$x$", fontsize=22); plt.ylabel("$y$", fontsize=22)
plt.xticks(np.linspace(-1, 1, 5), fontsize=14);
plt.yticks(np.linspace(-0.5, 2, 6), fontsize=14);
plt.gcf().subplots_adjust(bottom=0.15); plt.tight_layout();
plt.legend()

fig = mpl.pyplot.gcf()
fig.set_size_inches(8, 8)
```

### Iterative gradient-free optimisation

```python
Nms = np.array( [4, 12] ).astype(int)
Nqs = Nms*10

m_nu_s = 200

x_start = np.concatenate( [e_xa, [0]])
x_end = np.concatenate( [e_xc, [0]])

m0_path = pyritz.funcs.CollocationFF.get_straight_line_path(x_start, x_end, Nms[0],
                                    exclude_start_point=True, exclude_end_point=True)
m0_endpoint_rot = np.array([x_start[-1], x_end[-1]])
m0 = np.concatenate([m0_endpoint_rot, m0_path])

function_family = pyritz.funcs.CollocationFF(Nms[0], dim, derivatives=1,
                               fixed_start_point=x_start,
                               fixed_end_point=x_end)
quad_scheme = pyritz.quads.Q_clenshaw_curtis

compute_gradient=False

nlopt_algorithm = nlopt.LN_NEWUOA
maxeval = 5000
xtol_rel = 1e-12

m = m0

arch = []
dim = np.size(x_start)
prev_Nm = Nms[0]
for Nm, Nq in zip(Nms, Nqs):
    
    Nm, Nq = int(Nm), int(Nq)
    
    m_path = function_family.change_function_order(m[2:], prev_Nm, Nm, dim)
    m = np.concatenate([m[:2], m_path])
    function_family.set_order(Nm)

    m0_path = pyritz.funcs.CollocationFF.get_straight_line_path(x_start, x_end, Nm,
                                        exclude_start_point=True, exclude_end_point=True)
    m0_endpoint_rot = np.array([x_start[-1], x_end[-1]])
    m0 = np.concatenate([m0_endpoint_rot, m0_path])

    ff = pyritz.funcs.CollocationFF(Nm, dim, derivatives=1,
                                   fixed_start_point=x_start, fixed_end_point=x_end)

    quad_scheme = pyritz.quads.Q_clenshaw_curtis

    compute_gradient=False
    act = pyritz.Action(dim, ff, lagrangian, Nq, quad_scheme,
                        lagrangian_args=compute_gradient)

    def get_path(m):
        m_path = m[2:]
        x_start[-1] = m[0]
        x_end[-1] = m[1]

        ff.set_fixed_start_point(x_start)
        ff.set_fixed_end_point(x_end)
        return m_path

    def get_action(m, grad):
        return act.compute(get_path(m))

    opt = nlopt.opt(nlopt.LN_NEWUOA, np.size(m0))
    opt.set_min_objective(get_action)
    opt.set_xtol_rel(1e-12)
    opt.set_maxeval(5000)
    m = opt.optimize(m0)
    
    S = act.compute(get_path(m))
    
    arch.append( (m, S, Nm, Nq) )
    prev_Nm = Nm

    print("Nm:", Nm, " Nq:", Nq, " S:", S)
```

```python
# Plot path

ts = np.linspace(-1, 1, 1000)

paths = [ (m, "Nm=%d" % Nm, Nm) for m, S, Nm, Nq in arch ]

for p in paths:
    _m, _mlabel, _Nm = p
    _m = get_path(_m)
    xs, vs = ff.evaluate_at_order(_m, ts, _Nm)
    x, y, theta = xs[:, 0], xs[:, 1], xs[:, 2]
    px, py = np.array([np.cos(theta), np.sin(theta)])/10
    
    plt.plot(x, y, label=_mlabel)
    
    arrow_step = 30
    for i in range(len(px)):
        if i%arrow_step==0:
            plt.plot( [x[i],x[i]+px[i]], [y[i],y[i]+py[i]], color="black", alpha=0.6)

# Plot contour of potential

Nq=128
t=np.linspace(-1,1, Nq)
X, Y = np.meshgrid(np.linspace(-1.5, 1.2, 128), np.linspace(-.5, 2, 128))
plt.contour(X, Y, system_pot(X, Y), 256, cmap='gray', linewidths=1, alpha=0.4);

# Plot fixed points

plt.plot(e_xa[0], e_xa[1], 'o', mfc='#AA0000', mec='#AA0000', ms=4)
plt.plot(e_xb[0], e_xb[1], 'o', mfc='#AA0000', mec='#AA0000', ms=4)
plt.plot(e_xc[0], e_xc[1], 'o', mfc='#AA0000', mec='#AA0000', ms=4)
plt.plot(e_xs1[0], e_xs1[1], 'o', mfc='#00AA22', mec='#00AA22', ms=4)
plt.plot(e_xs2[0], e_xs2[1], 'o', mfc='#00AA22', mec='#00AA22', ms=4)

# Plot limits and setup

plt.xlim(-1.3, 1.2);plt.ylim(-.5, 2);
plt.xlabel("$x$", fontsize=22); plt.ylabel("$y$", fontsize=22)
plt.xticks(np.linspace(-1, 1, 5), fontsize=14);
plt.yticks(np.linspace(-0.5, 2, 6), fontsize=14);
plt.gcf().subplots_adjust(bottom=0.15); plt.tight_layout();
plt.legend()

fig = mpl.pyplot.gcf()
fig.set_size_inches(8, 8)
```

### Gradient optimisation

```python
m_nu_s = 100

x_start = np.concatenate( [e_xa, [0]])
x_end = np.concatenate( [e_xc, [0]])

Nm = 10
Nq = Nm*10

m0_path = pyritz.funcs.CollocationFF.get_straight_line_path(x_start, x_end, Nm,
                                    exclude_start_point=True, exclude_end_point=True)
m0_endpoint_rot = np.array([x_start[-1], x_end[-1]])
m0 = np.concatenate([m0_endpoint_rot, m0_path])

ff = pyritz.funcs.CollocationFF(Nm, dim, derivatives=1,
                               fixed_start_point=x_start, fixed_end_point=x_end)

quad_scheme = pyritz.quads.Q_clenshaw_curtis

compute_gradient=True
act = pyritz.Action(dim, ff, lagrangian, Nq, quad_scheme,
                    lagrangian_args=compute_gradient)

def get_path(m):
    m_path = m[2:]
    x_start[-1] = m[0]
    x_end[-1] = m[1]
    
    ff.set_fixed_start_point(x_start)
    ff.set_fixed_end_point(x_end)
    return m_path

def get_action(m, grad):
    m_path = get_path(m)
    
    if grad.size > 0:
        grad[0] = 0
        grad[1] = 0
        
        return act.compute(m_path, grad[2:])
    else:
        return act.compute(m_path)

opt = nlopt.opt(nlopt.LD_MMA, np.size(m0))
opt.set_min_objective(get_action)
opt.set_xtol_rel(1e-12)
opt.set_maxeval(1000)
m = opt.optimize(m0)

print(act.compute(get_path(m0)))
print(act.compute(get_path(m)))
```

```python
# Plot path

ts = np.linspace(-1, 1, 1000)

paths = [
    (m0, "Initial", "blue"),
    (m, "Final", "green")
]

for p in paths:
    _m, _mlabel, _color = p
    _m = get_path(_m)
    xs, vs = ff.evaluate(_m, ts)
    x, y, theta = xs[:, 0], xs[:, 1], xs[:, 2]
    px, py = np.array([np.cos(theta), np.sin(theta)])/10
    
    plt.plot(x, y, label=_mlabel, color=_color)
    
    arrow_step = 30
    for i in range(len(px)):
        if i%arrow_step==0:
            plt.plot( [x[i],x[i]+px[i]], [y[i],y[i]+py[i]], color="black", alpha=0.6)

# Plot contour of potential

Nq=128
t=np.linspace(-1,1, Nq)
X, Y = np.meshgrid(np.linspace(-1.5, 1.2, 128), np.linspace(-.5, 2, 128))
plt.contour(X, Y, system_pot(X, Y), 256, cmap='gray', linewidths=1, alpha=0.4);

# Plot fixed points

plt.plot(e_xa[0], e_xa[1], 'o', mfc='#AA0000', mec='#AA0000', ms=4)
plt.plot(e_xb[0], e_xb[1], 'o', mfc='#AA0000', mec='#AA0000', ms=4)
plt.plot(e_xc[0], e_xc[1], 'o', mfc='#AA0000', mec='#AA0000', ms=4)
plt.plot(e_xs1[0], e_xs1[1], 'o', mfc='#00AA22', mec='#00AA22', ms=4)
plt.plot(e_xs2[0], e_xs2[1], 'o', mfc='#00AA22', mec='#00AA22', ms=4)

# Plot limits and setup

plt.xlim(-1.3, 1.2);plt.ylim(-.5, 2);
plt.xlabel("$x$", fontsize=22); plt.ylabel("$y$", fontsize=22)
plt.xticks(np.linspace(-1, 1, 5), fontsize=14);
plt.yticks(np.linspace(-0.5, 2, 6), fontsize=14);
plt.gcf().subplots_adjust(bottom=0.15); plt.tight_layout();
plt.legend()

fig = mpl.pyplot.gcf()
fig.set_size_inches(8, 8)
```
