```python
def get_chebyshev_diff_matrix(N, D_ord):
    _, Ds = dmsuite.chebdif(N, D_ord)
    
    # Required because we use a different node ordering
    for i in range(D_ord):
        Ds[i, :, :] = ((-1)**(i+1)) * Ds[i, :, :]
    return Ds

def generate_barycentric_weights(Nm):
    """Generates the barycentric weights of order *Nm*.

    :type Nm: int
    """
    j = np.arange(0, Nm)
    w = np.power(-1, j)
    o = np.ones(Nm)
    o[0] = 0.5
    o[Nm-1] = 0.5
    w = np.flipud(w*o)

    return w

def get_chebyshev_nodes(N):
    """Returns a Chebyshev grid with *N* nodes.

    :type N: int
    :rtype: np.array
    """
    ts = -np.cos(np.pi * np.arange(N) / (N-1)) # Minus sign here is so that the the points are in increasing order
    return ts


def barycentric_1d(N, Nq, quad_scheme, deriv_orders):
    qxs, qw = quad_scheme(Nq)
    xs = get_chebyshev_nodes(N)
    wx = generate_barycentric_weights(N)
    
    if len(deriv_orders) == 0:
        raise Exception("deriv_orders can't be empty.")
    
    if np.max(deriv_orders) != 0:
        Ds = get_chebyshev_diff_matrix(N, np.max(deriv_orders))
        
    A =  np.zeros( (Nq, N) )
    for i in range(Nq):
        delta_x = qxs[i] - xs
        whr = np.where(delta_x == 0)[0]

        if len(whr) == 0:
            wd = wx/delta_x
            A[i, :] = wd / np.sum(wd)
        else:
            A[i, whr] = 1
    
    Bs = []
    
    for i in range(len(deriv_orders)):
        if deriv_orders[i] == 0:
            Bs.append(A)
        else:
            D = Ds[deriv_orders[i]-1, :, :]
            Bs.append(A.dot(D))

    return Bs
```

## n-dimensional Barycentric formula

```python
def barycentric_nd(dim, Ns, Nqs, quad_scheme, deriv_orders):    
    deriv_orders = np.array(deriv_orders)
    
    if len(Ns) != dim:
        raise Exception("Nxs must match dim.")
    if len(Nqs) != dim:
        raise Exception("Nxs must match dim.")
    if deriv_orders.shape[1] != dim:
        raise Exception("The elements of deriv_orders must match dim.")
    
    nBs = []
    
    for i in range(dim):
        Bs = barycentric_1d(Ns[i], Nqs[i], quad_scheme, deriv_orders[:, i])
        nBs.append(Bs)
        
    krBs = []
    for i in range(deriv_orders.shape[0]):
        krB = 1
        
        for j in range(dim):
            krB = np.kron(krB, nBs[j][i])
            
        krBs.append(krB)
            
    if len(krBs) == 1:
        return krBs[0]
    else:
        return krBs
```

## Ginzberg-Landau system

```python
def get_action(_m, grad):
    opt.evals +=1
    m_t.reshape( (Nt, Nx) )[1:-1, 1:-1] = _m.reshape( (Nt-2, Nx-2) )
    
    U[:] = A.dot(m_t)
    Ut[:] = Bt.dot(m_t)
    Uxx[:] = Bxx.dot(m_t)
    
    p[:] = 2*m_r_T*Ut - 4*m_delta*Uxx - m_r_delta*U*( 1 - U**2 )
    ls[:] = 0.25*m_T * ( p )**2
    S = qw.dot(ls)
    
    if not grad is None and grad.size > 0:
        opt.gevals += 1
        #dL/du
        dls[0, :] = 0.5*m_r_T*m_r_delta*p*(3*U**2-1)
        #dL/dut
        dls[1, :] = 0.25*p
        #dL/duxx
        dls[2, :] = -2*m_delta*m_T*p
        
        grad[:] = np.einsum("dj, dji", dls, C).reshape( (Nt, Nx) )[1:-1, 1:-1].flatten()

    #if opt.evals % 10 == 0:
    #    print(opt.evals)
        
    return S

def uplus(xs, delta):
    return np.tanh( 0.5*(1 - np.abs(xs))/ ( np.sqrt(2) * delta ))

def u0(T, X, delta):
    return -T*np.tanh( 0.5*(1 - np.abs(X))/ ( np.sqrt(2) * delta ))


def u0_t(T, X, delta):
    return -np.tanh((1 - np.sqrt(X**2))/(2.*np.sqrt(2)*m_delta))

def u0_xx(T, X, delta):
    return         (-2*T*csch((-1 + np.sqrt(X**2))/(np.sqrt(2)*m_delta))**3*np.sinh((-1 + np.sqrt(X**2))/(2.*np.sqrt(2)*m_delta))**4)/m_delta**2
```
