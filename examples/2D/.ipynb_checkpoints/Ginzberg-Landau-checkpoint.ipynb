### 2D  Barycentric Interpolation test

```python
def get_chebyshev_nodes(N):
    """Returns a Chebyshev grid with *N* nodes.

    :type N: int
    :rtype: np.array
    """
    ts = -np.cos(np.pi * np.arange(N) / (N-1)) # Minus sign here is so that the the points are in increasing order
    return ts

def generate_barycentric_weights(Nm):
    """Generates the barycentric weights of order *Nm*.

    :type Nm: int
    """
    j = np.arange(0, Nm)
    w = np.power(-1, j)
    o = np.ones(Nm)
    o[0] = 0.5
    o[Nm-1] = 0.5
    w = np.flipud(w*o)

    return w

def setup_barycentric_2D(Nx, Ny, Nqx, Nqy):
    # A and B matrices
    
    _A = np.zeros( (Nqx, Nqy, Nx, Ny) )
    A = np.zeros( (Nqx*Nqy, Nx*Ny) )
    Ax = np.zeros( (Nqx*Nqy, Nx*Ny) )
    Ayy = np.zeros( (Nqx*Nqy, Nx*Ny) )
    
    # Get Chebyshev differentiation matrices
    
    Dx = dmsuite.chebdif(Nx, 1)[1][0, :, :]
    Dyy = dmsuite.chebdif(Ny, 2)[1][1, :, :]
    
    # Construct the weight matrix

    wxs = generate_barycentric_weights(Nx)
    wys = generate_barycentric_weights(Ny)
    W = np.outer(wxs, wys)

    # Construct Chebsyhev grid for the collocation

    cxs = -np.cos(np.pi * np.arange(Nx) / (Nx-1))
    cys = -np.cos(np.pi * np.arange(Ny) / (Ny-1))
    cXY = np.outer(cxs, cys)
    
    # Chebyshev points for the quadrature
    
    qxs = -np.cos(np.pi * np.arange(Nqx) / (Nqx-1))
    qys = -np.cos(np.pi * np.arange(Nqy) / (Nqy-1))
    
    for i in range(Nqx):
        for j in range(Nqy):
            qx, qy = qxs[i], qys[j]
            deltax = qx - cxs
            deltay = qy - cys
            delta = np.outer( deltax, deltay )
            
            # See if collocation point is equal quadrature node
            whrx = np.where(deltax == 0)[0]
            whry = np.where(deltay == 0)[0]
            
            if len(whrx) == 0 and len(whry)==0:
                wd = W/delta
                _A[i, j, :, :] = wd / np.sum(wd)
            elif len(whrx)==1 and len(whry)==0:
                wd = W[whrx, :].flatten()/deltay
                _A[i, j, whrx, :] = wd / np.sum(wd)
            elif len(whrx)==0 and len(whry)==1:
                wd = W[:, whry].flatten()/deltax
                _A[i, j, :, whry] = wd / np.sum(wd)
            elif len(whrx)==1 and len(whry)==1:
                _A[i, j, whrx, whry] = 1
            else:
                raise Exception("Something wrong here.")
                
            A[Nqy*i + j, :] = _A[i, j, :, :].flatten()
            Ax[Nqy*i + j, :] = np.einsum("kl,kr", _A[i, j, :, :], Dx).T.flatten()
            Ayy[Nqy*i + j, :] = np.einsum("kl,lr", _A[i, j, :, :], Dyy).flatten()
            
    return A, Ax, Ayy

def get_action(_m, grad):
    m.reshape( (Nt+2, Nx+2) )[1:-1, 1:-1] = _m.reshape( (Nt, Nx) )
    
    U = A.dot(m0)
    Ut = At.dot(m0)
    Uxx = Axx.dot(m0)
    
    ls[:] = 0.25*m_T * ( 0.5*m_T*Ut - 4*m_delta*Uxx - m_r_delta*U*( 1 - U**2 ) )**2
    
    return qw.dot(ls)
    
m_T = 1
m_delta = 0.03
    
Nx = 24
Nt = 24
Nqx = Nx*4
Nqt = Nt*4

m = np.zeros( (Nt+2, Nx+2) ).flatten()
ls = np.zeros( Nqt*Nqx )

quad_scheme = pyritz.quads.Q_clenshaw_curtis

qw = np.outer(quad_scheme(Nqt)[1], quad_scheme(Nqx)[1]).flatten()

A, At, Axx = setup_barycentric_2D(Nt+2, Nx+2, Nqt, Nqx)

u_start = uplus(pyritz.funcs.CollocationFF.get_chebyshev_nodes(Nx+2), 0.03)
u_end = -uplus(pyritz.funcs.CollocationFF.get_chebyshev_nodes(Nx+2), 0.03)
m0 = pyritz.funcs.CollocationFF.get_straight_line_path(u_start, u_end, Nt+2)
m0 = m0.reshape(Nx+2, Nt+2).T.flatten()

U = A.dot(m0)
Ut = At.dot(m0)
Uxx = Axx.dot(m0)

_m = m0.reshape(Nt+2, Nx+2)[1:-1, 1:-1]
get_action(_m, None)
```

### Fixed points

```python
def uplus(xs, delta):
    return np.array(
        [ np.tanh( 0.5*(1 + x)/ ( np.sqrt(2) * delta ) ) if x < 0 else
          np.tanh( 0.5*(1 - x)/ ( np.sqrt(2) * delta ) ) for x in xs]
    )

def duplus(xs, delta):
    return np.array(
        [ np.sech( 0.5*(1 + x)/ ( np.sqrt(2) * delta ) )**2 / (2 * (np.sqrt(2) * m_delta)) if x < 0 else
          -np.tanh( 0.5*(1 - x)/ ( np.sqrt(2) * delta ) )**2/ (2 * (np.sqrt(2) * m_delta)) for x in xs]
    )
m0_full
def dduplus(xs, delta):
    return np.array(
        [ -2*np.cosh( 0.5*(1 + x)/ ( np.sqrt(2) * delta ) )**(-2) * np.tanh( 0.5*(1 + x)/ ( np.sqrt(2) * delta ) ) / ( (2 * (np.sqrt(2) * m_delta))**2 ) if x < 0 else
          -2*np.cosh( 0.5*(1 - x)/ ( np.sqrt(2) * delta ) )**(-2) * np.tanh( 0.5*(1 - x)/ ( np.sqrt(2) * delta ) ) / ( (2 * (np.sqrt(2) * m_delta))**2 ) for x in xs]
    )


Nu = 20
ff = pyritz.funcs.CollocationFF(Nu, 1, derivatives=0)
x_start = uplus(pyritz.funcs.CollocationFF.get_chebyshev_nodes(Nu), 0.03)
plt.plot(ff.evaluate(x_start, np.linspace(-1, 1, 1000)))
```
