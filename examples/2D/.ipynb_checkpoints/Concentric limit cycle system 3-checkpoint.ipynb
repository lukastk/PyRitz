# Concentric Limit Cycle System

**Lagrangian:** Geometric Action
<br>
**Implementation:** Python

**System:**

$ \begin{cases}
du = (u - u^3 - \beta u v^2)\ dt + \sqrt{\epsilon}\ dW_u \\
dv = -(1 + u^2) v\ dt + \sqrt{\epsilon}\ dW_v
\end{cases}$

## System

### Parameters

System parameters are defined here. All system parameter variables are prefaced with "m_", with the exception of `dim` which is the dimension of the system.

```python
dim = 2

m_r1 = 1
m_r2 = 3
m_r3 = 5
m_c = 0.25
```

### Dynamics

`system_a` is the drift term of the system. It's used for plotting stream-plots, as well as finding fixed points of the deterministic system.

`lagrangian` is the Lagrangian of the system. The Lagrangian used is the Geometric Action lagrangian (Heymann and Vanden-Eijnden), defined as

$$
L(x, \dot x) = |\dot x| |a(x)| - \dot x \cdot a(x)
$$

for any general Ito diffusion of the form

$$
dX = a(X)dt + \sqrt{\epsilon} dW.
$$

The function also computes the partial derivatives of the Lagrangian, which are used for gradient optimisation. This can be enabled/disabled by setting the *args* argument to True/False. The partial derivatives of the geometric lagrangian are

$$\begin{aligned}
\frac{\partial L}{\partial x} & = \nabla a( x ) \cdot \left( |\dot x| \hat{a}(x) - \dot x \right) \\
\frac{\partial L}{\partial \dot x} & = |a(x)| \hat{\dot{x}} - a(x)
\end{aligned}$$

An important property of the geometric action is that it is time-reparameterisation invariant, meaning that $x$ can be replaced by any other path that shares the same *graph*.

#### Notes on the Maier-Stein system

Let $(\nabla a)_{ij} = \frac{\partial a_i}{\partial x_j}$, where $x = (u, v)$, then

$$
\nabla a = \begin{pmatrix}
    1 - 3 u^2 - \beta v^2 &  -2 u v \\
    - 2 \beta u v & - (1 + u^2)
\end{pmatrix}
$$

```python
def system_a(r, th):
    return np.array([
        -m_c*(1 + np.cos(th)**2)*(r - m_r1)*(r - m_r2)*(r - m_r3)*r,
        r*(1 + np.sin(th)**2)
    ])

def system_a_cartesian(x, y):
    return np.array([
        -((y*(x**2 + 2*y**2))/np.sqrt(x**2 + y**2)) + (x*(-1 - x**2/(x**2 + y**2))*(-4 + np.sqrt(x**2 + y**2))*(-3 + np.sqrt(x**2 + y**2))*(-1 + np.sqrt(x**2 + y**2)))/4.,
x*np.sqrt(x**2 + y**2)*(1 + y**2/(x**2 + y**2)) + (y*(-1 - x**2/(x**2 + y**2))*(-4 + np.sqrt(x**2 + y**2))*(-3 + np.sqrt(x**2 + y**2))*(-1 + np.sqrt(x**2 + y**2)))/4.
    ])

def lagrangian(ls, dxls, dvls, fvals, ts, args):
    compute_gradient = args
    
    xs, dxs = fvals
    rs, ths = xs[:, 0], xs[:, 1]
    drs, dths = dxs[:, 0], dxs[:, 1]
    
    ars = -m_c*(1+np.cos(ths)**2)*(rs - m_r1)*(rs - m_r2)*(rs - m_r3)*rs
    aths = rs*(1+np.sin(ths)**2)
    aas = np.array( [ars, aths] )
    
    dxs_norm = np.linalg.norm(dxs, axis=1)
    a_norm = np.sqrt(ars**2 + aths**2)
    dxs_dot_a = ars*dxs[:,0] + aths*dxs[:, 1]
    
    ls[:] = dxs_norm * a_norm - dxs_dot_a
    
    # Compute gradient
    
    if compute_gradient:

        ## Compute dL/dx

        da = np.zeros( (2, 2, len(ts)) )
        da[0,0] = -((4*rs**3 - m_r1*m_r2*m_r3 - 3*rs**2*(m_r1 + m_r2 + m_r3) + 2*rs*(m_r2*m_r3 + m_r1*(m_r2 + m_r3)))*(3 + np.cos(2*ths)))/8.
        da[0,1] = 1 + np.sin(ths)**2
        da[1,0] = (rs*(rs - m_r1)*(rs - m_r2)*(rs - m_r3)*np.sin(2*ths))/4.
        da[1,1] = rs*np.sin(2*ths)
        
        a_normalized = aas/a_norm

        a_normalized[:, np.where(a_norm==0)] = 0 # Set zero-norm vectors to 0

        for i in range(dxls.shape[0]):
            dxls[i, :] = da[:,:,i].dot( dxs_norm[i]*a_normalized[:,i] - dxs[i,:])

        ## Compute dL/dv

        dxs_2 = (dxs.T*(a_norm/dxs_norm)).T
        dxs_2[ np.where(dxs_norm==0),:] = 0

        dvls[:] = dxs_2 - aas.T
```

```python
Nms = np.arange(4, 20, 2)
Nqs = Nms*10

x_start = np.array([1, 0])
x_end = np.array([3, 20])
x_f = 1.3

m0 = pyritz.funcs.CollocationFF.get_straight_line_path(x_start, x_end, Nms[0],
                                    exclude_start_point=True, exclude_end_point=True)

arch = []

prev_Nm = Nms[0]

for Nm, Nq in zip(Nms, Nqs):
    
    quad_scheme = pyritz.quads.Q_clenshaw_curtis
    qts, qws = quad_scheme(Nq)
    
    prev_ff = pyritz.funcs.CollocationFF(dim, fixed_start_point=x_start, fixed_end_point=x_end)
    prev_ff.initialise(Nm, qts)
    
    prev_ff = pyritz.funcs.CollocationFF(prev_Nm, qts, dim, deriv_orders=1,
                               fixed_start_point=x_start,
                               fixed_end_point=x_end)
    ts = pyritz.funcs.CollocationFF.get_chebyshev_nodes(Nm+2)
    m0 = prev_ff.evaluate_at(m0, ts)[0][1:-1,:].T.flatten()
    m0[Nm:] = x_f*m0[Nm:]
    m0[Nm] = 0
    #x_end[1] *= x_f
    #x_end[1] += 10
    
    m0 = pyritz.funcs.CollocationFF.get_straight_line_path(x_start, x_end, Nm,
                                        exclude_start_point=True, exclude_end_point=True)

    ff = pyritz.funcs.CollocationFF(Nm, qts, dim, deriv_orders=1,
                               fixed_start_point=x_start,
                               fixed_end_point=x_end)
    
    compute_gradient=True
    act = pyritz.Action(dim, ff, lagrangian, Nq, quad_scheme, lagrangian_args=compute_gradient)

    def get_action(m, grad):
        if grad.size > 0:
            return act.compute(m, grad)
        else:
            return act.compute(m)

    opt = nlopt.opt(nlopt.LD_SLSQP, np.size(m0))
    opt.set_min_objective(get_action)
    opt.set_xtol_rel(1e-12)
    #opt.set_maxeval(5000)
    m0 = opt.optimize(m0)
    
    #S_acc = act.compute_at_order(m0, Nq*20)
    S_acc = act.compute(m0)
    arch.append( (m0, S_acc) )
    
    print("Nm:", Nm, "S_acc:", S_acc, "x_end[1]:", x_end[1])
    
    prev_Nm = Nm
```

```python
Nms = np.arange(2, 40, 1)
Nqs = Nms*10

x_start = np.array([1, 0])
x_end = np.array([3, 20])
x_f = 1.5

m0 = pyritz.funcs.CollocationFF.get_straight_line_path(x_start, x_end, Nms[0])

arch = []

prev_Nm = Nms[0]

for Nm, Nq in zip(Nms, Nqs):
    
    prev_ff = pyritz.funcs.CollocationFF(prev_Nm, dim, derivatives=1)
    ts = pyritz.funcs.CollocationFF.get_chebyshev_nodes(Nm)
    m0 = prev_ff.evaluate(m0, ts)[0].T.flatten()
    m0[Nm:] = x_f*m0[Nm:]
    m0[Nm] = 0
    x_start[1] = m0[Nm]
    x_end[1] = m0[-1]
    
    m0[1:Nm-1] += np.random.random(Nm-2)*0.00001
    m0[Nm+1:-1] += np.random.random(Nm-2)*0.00001
    
    ff = pyritz.funcs.CollocationFF(Nm, dim, derivatives=1)

    quad_scheme = pyritz.quads.Q_clenshaw_curtis

    compute_gradient=True
    act = pyritz.Action(dim, ff, lagrangian, Nq, quad_scheme, lagrangian_args=compute_gradient)

    ub = np.full(dim*Nm, np.inf)
    lb = np.full(dim*Nm, -np.inf)

    tol = 1e-10

    for i in range(dim):
        ub[i*Nm] = x_start[i] + tol
        lb[i*Nm] = x_start[i] - tol
        ub[Nm-1 + i*Nm] = x_end[i] + tol
        lb[Nm-1 + i*Nm] = x_end[i] - tol

    ub[2*Nm-1]= np.inf
    lb[2*Nm-1]= -np.inf

    def get_action(m, grad):
        if grad.size > 0:
            grad[0] = grad[Nm] = grad[Nm-1] = np.inf 
            return act.compute(m, grad)
        else:
            return act.compute(m)

    opt = nlopt.opt(nlopt.LD_SLSQP, np.size(m0))
    opt.set_lower_bounds(lb)
    opt.set_upper_bounds(ub)
    opt.set_min_objective(get_action)
    opt.set_xtol_rel(1e-12)
    #opt.set_maxeval(5000)
    m0 = opt.optimize(m0)
    
    S_acc = act.compute_at_order(m0, Nq*20)
    arch.append( (m0, S_acc) )
    
    print("Nm:", Nm, "S_acc:", S_acc, "x_end[1]:", m0[-1])
    
    prev_Nm = Nm
```

```python
x_start = np.array([1, 0])
x_end = np.array([3, 20])

Nm = 4
Nq = Nm*10

m0 = pyritz.funcs.CollocationFF.get_straight_line_path(x_start, x_end, Nm)
#m0 = np.array(m2)
#m0[Nm:] = 4*m0[Nm:]
m0[Nm] = 0
x_start[1] = m0[Nm]

quad_scheme = pyritz.quads.Q_clenshaw_curtis
qts, qws = quad_scheme(Nq)

ff = pyritz.funcs.CollocationFF(dim, fixed_start_point=x_start, fixed_end_point=x_end)
ff.initialise(Nm, qts)

compute_gradient=True
act = pyritz.Action(dim, ff, lagrangian, Nm, Nq, quadrature_scheme=quad_scheme,
                    lagrangian_args=compute_gradient)

ub = np.full(dim*Nm, np.inf)
lb = np.full(dim*Nm, -np.inf)

tol = 1e-10

for i in range(dim):
    ub[i*Nm] = x_start[i] + tol
    lb[i*Nm] = x_start[i] - tol
    ub[Nm-1 + i*Nm] = x_end[i] + tol
    lb[Nm-1 + i*Nm] = x_end[i] - tol
    
ub[2*Nm-1]= np.inf
lb[2*Nm-1]= -np.inf

def get_action(m, grad):
    if grad.size > 0:
        grad[0] = grad[Nm] = grad[Nm-1] = np.inf 
        return act.compute(m, grad)
    else:
        return act.compute(m)

opt = nlopt.opt(nlopt.LD_SLSQP, np.size(m0))
opt.set_lower_bounds(lb)
opt.set_upper_bounds(ub)
opt.set_min_objective(get_action)
opt.set_xtol_rel(1e-12)
opt.set_maxeval(5000)
m = opt.optimize(m0)

print(act.compute(m0))
print(act.compute(m))
```
